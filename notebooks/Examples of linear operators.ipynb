```python
from epsilon.problems import benchmark
from epsilon.compiler import compiler
from epsilon import cvxpy_expr
from epsilon import tree_format
from epsilon.expression_pb2 import Expression
```

```python
def get_prox(expr):
    if expr.expression_type == Expression.ADD:
        return get_prox(expr.arg[0])
    if expr.expression_type == Expression.MULTIPLY:
        return get_prox(expr.arg[1])
    return expr
```

```python
for problem in benchmark.PROBLEMS:
    cvxpy_prob = problem.create()
    print problem.name
    print "-----"
    prob = compiler.compile_problem(cvxpy_expr.convert_problem(cvxpy_prob))
    for arg in prob.objective.arg:
        prox = get_prox(arg)
        if prox.proximal_operator.name == "LeastSquaresProx":
            print "LeastSquares"
            print tree_format.format_expr(prox.arg[0].arg[0])
        if prox.proximal_operator.name == "LinearEqualityProx":
            print "LinearEquality"
            print tree_format.format_expr(prox.arg[0])
    print
```
